### Unsorted List

### Function to check whether the list is sorted or unsorted.

```python
def is_sorted(lst):
  for pos in range(len(lst)-1):
    curr_element=lst[pos]
    nxt_element=lst[pos+1]
    if curr_element>nxt_element:
      print("The list is unsorted")
      return False
  print("The list is sorted")
  return True
```

```python
is_sorted(unsorted_list)
```

### Bubble sort

```python
def bubble_sort(lst):
  print("The given list =", lst)
  # print("----------------------------")
  iteration=0
  while is_sorted(lst) is False:
    iteration=iteration+1
    step=0
    for i in range(len(lst)-1):
      step=step+1
      curr_element=lst[i]
      nxt_element=lst[i+1]
      if curr_element>nxt_element:
        lst[i],lst[i+1]=lst[i+1],lst[i]


      print("\t Step ",step, lst)
    print("----------------------------")
    print(f"The List after iteration {iteration} ={lst}")

  return lst

```

```python
unsorted_list=[1,9,4,3,2,7]
bubble_sort(unsorted_list)
```

### Insertion Sort

```python
def insertion_sort(lst):
  print("The given list =", lst)
  print("----------------------------")
  for pos in range(len(lst)):
    ref_pos=0
    while pos > ref_pos:
      if lst[pos]<lst[ref_pos]:
        lst.insert(ref_pos,lst[pos])
        lst.pop(pos+1)
        break
      ref_pos=ref_pos+1
    print(f"The List after iteration {pos+1} ={lst}")

    if is_sorted(lst)== True:
      break
```

```python
unsorted_list=[1,9,4,3,2,7]
insertion_sort(unsorted_list)
```

### Selection Sort

```python
def selection_sort(lst):
  print("The given list =", lst)
  print("----------------------------")

  for pos in range (1,len(lst)-1):
    ref_element=lst[pos]
    sub_list=lst[pos:]
    min_sublist=min(sub_list)
    lst[pos]=min_sublist
    lst[pos+sub_list.index(min_sublist)]=ref_element
    print(lst)

    if is_sorted(lst)== True:
      break

```

```python
unsorted_list=[1,9,4,3,2,7]
selection_sort(unsorted_list)
```
